## Step 1: Project context

FloodSense Lokoja is a flood-risk intelligence and decision-support workflow for Lokoja, Kogi State, Nigeria. The system combines terrain, hydrology, rainfall, land cover, Sentinel-1 SAR validation, exposure analysis, and priority ranking.

The first rule of this project is to keep the analysis focused on **one study area: Lokoja**. Do not expand to national-scale modelling until this workflow is strong and reproducible.

```python
from pathlib import Path

PROJECT_ROOT = Path.cwd()
print('Current working directory:', PROJECT_ROOT)
print('Tip: run notebooks from the repository root so relative paths work correctly.')
```

## Step 2: Load the project configuration

The config file stores project metadata, paths, CRS, model weights, hydrology settings, SAR settings, and rainfall scenario modifiers. This keeps notebooks and scripts from hardcoding paths.

```python
from floodsense.config import load_config

config = load_config()
config['project']
```

## Step 3: Confirm the target CRS

The default projected CRS is `EPSG:32632`, which is suitable for metric distance and area calculations around Lokoja. Distance-to-stream, road-length, and area summaries should use a projected CRS, not latitude/longitude.

```python
target_crs = config['project'].get('crs', 'EPSG:32632')
print('Target analysis CRS:', target_crs)
```

## Step 4: Ensure project directories exist

This creates the expected raw, interim, processed, output, and dashboard folders. It does not download or create fake GIS data.

```python
from floodsense.paths import ensure_project_directories

ensure_project_directories(config)
print('Project directories checked/created.')
```

## Step 5: Inspect expected raw data folders

Before running analysis, check which raw datasets are already present. Boundary data is required for preprocessing. Other datasets can be added gradually.

```python
from floodsense.paths import list_input_files

for key, folder in config['paths'].items():
    if key.startswith('raw_'):
        files = list_input_files(folder)
        print(f'{key:20s} -> {len(files)} file(s)')
        for file in files[:5]:
            print('   ', file.name)
        if len(files) > 5:
            print('    ...')
```
