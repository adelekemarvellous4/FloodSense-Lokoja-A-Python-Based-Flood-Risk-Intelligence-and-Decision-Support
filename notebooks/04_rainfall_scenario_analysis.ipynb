# 04 Rainfall Scenario Analysis

This notebook creates low, moderate, and severe flood scenario rasters from the susceptibility score.

## What you will do

1. Review rainfall scenario modifiers.
2. Confirm the susceptibility score exists.
3. Run scenario generation.
4. Inspect scenario outputs.
5. Note future return-period improvements.

## Step 1: Load configuration and scenario modifiers

```python
from pathlib import Path
import pandas as pd
from floodsense.config import load_config
from floodsense.rainfall import create_scenario_modifiers

config = load_config()
paths = config['paths']
modifiers = create_scenario_modifiers(config)
pd.DataFrame([{'scenario': k, 'modifier': v} for k, v in modifiers.items()])
```

## Step 2: Confirm susceptibility score input

Scenario generation starts from `susceptibility_score.tif`. If it is missing, run notebook 03 first.

```python
score_path = Path(paths['processed_rasters']) / 'susceptibility_score.tif'
print('Susceptibility score:', score_path)
print('Exists:', score_path.exists())
```

## Step 3: Run rainfall scenario workflow

The starter scenario model multiplies the susceptibility score by each rainfall modifier and clips the result to 0-1.

This is a planning scenario tool, not yet a return-period hydrological model.

```python
from floodsense.rainfall import run_rainfall_scenarios

if score_path.exists():
    run_rainfall_scenarios(config)
else:
    print('Skipped scenarios because susceptibility_score.tif is missing.')
```

## Step 4: Inspect scenario outputs

```python
scenario_files = []
for scenario in ['low', 'moderate', 'severe']:
    path = Path(paths['processed_rasters']) / f'flood_scenario_{scenario}.tif'
    scenario_files.append({'scenario': scenario, 'path': path, 'exists': path.exists()})
pd.DataFrame(scenario_files)
```

## Step 5: Create scenario preview maps

```python
from floodsense.mapping import plot_raster_preview

for row in scenario_files:
    if row['exists']:
        plot_raster_preview(row['path'], Path(paths['outputs_maps']) / f"flood_scenario_{row['scenario']}_preview.png", f"{row['scenario'].title()} Flood Scenario")
        print('Preview saved:', row['scenario'])
```
