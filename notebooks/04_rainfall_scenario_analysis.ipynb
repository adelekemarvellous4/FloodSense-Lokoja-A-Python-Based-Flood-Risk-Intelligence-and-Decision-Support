# 04 Rainfall Scenario Analysis

This notebook creates low, moderate, and severe flood scenario rasters by modifying the susceptibility score.

## Scenario Concept

The starter scenario model uses simple modifiers. Future versions should use rainfall return periods such as 10-year, 50-year, and 100-year events.

```python
from floodsense.config import load_config
from floodsense.rainfall import run_rainfall_scenarios

config = load_config()
run_rainfall_scenarios(config)
```

```python
from floodsense.mapping import plot_raster_preview
from pathlib import Path

rasters = Path(config['paths']['processed_rasters'])
maps = Path(config['paths']['outputs_maps'])
for scenario in ['low', 'moderate', 'severe']:
    plot_raster_preview(rasters / f'flood_scenario_{scenario}.tif', maps / f'flood_scenario_{scenario}_preview.png', f'{scenario.title()} Scenario')
```
