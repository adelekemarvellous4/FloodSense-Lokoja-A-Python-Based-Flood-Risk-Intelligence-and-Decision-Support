# 07 Priority Index and Decision Support

This notebook turns flood hazard and exposure information into a Flood Intervention Priority Index.

## What you will do

1. Load the exposure summary table.
2. Understand priority indicators.
3. Normalize available indicators.
4. Compute priority scores.
5. Classify areas as Low, Moderate, High, or Critical priority.

## Step 1: Load configuration and exposure summary

```python
from pathlib import Path
import pandas as pd
from floodsense.config import load_config

config = load_config()
paths = config['paths']
exposure_path = Path(paths['processed_tables']) / 'exposure_summary.csv'
print('Exposure summary:', exposure_path)
print('Exists:', exposure_path.exists())
if exposure_path.exists():
    display(pd.read_csv(exposure_path))
```

## Step 2: Understand priority indicators

The priority index uses available columns only. Expected indicators include:

- `exposed_population`
- `exposed_buildings`
- `affected_road_length_km`
- `hazard_score` if available

Missing indicators are reported and skipped. The model does not invent values.

## Step 3: Run priority index workflow

Outputs:

- `data/processed/tables/priority_ranking.csv`
- `outputs/tables/priority_ranking.csv`

```python
from floodsense.priority_index import run_priority_index

if exposure_path.exists():
    run_priority_index(config)
else:
    print('Skipped priority index because exposure_summary.csv is missing.')
```

## Step 4: Review priority ranking

```python
ranking_path = Path(paths['processed_tables']) / 'priority_ranking.csv'
if ranking_path.exists():
    ranking = pd.read_csv(ranking_path)
    display(ranking)
else:
    print('Priority ranking not available yet.')
```

## Step 5: Inspect class counts

```python
if ranking_path.exists() and 'priority_class' in ranking.columns:
    display(ranking['priority_class'].value_counts().rename_axis('priority_class').reset_index(name='count'))
else:
    print('Priority class counts not available.')
```

## What to check before moving on

- Should results be aggregated by ward/community polygons?
- Are vulnerability indicators needed, such as access to shelters or distance to roads?
